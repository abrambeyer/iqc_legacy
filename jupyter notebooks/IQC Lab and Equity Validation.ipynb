### Check completeness files for existence of our hospitals

```python
result_frame = []
for i, item in enumerate(os.listdir(filepath)):
    if (item.endswith('.zip')  == False) and (item.endswith('xlsx') == False) and (item != 'archive'):
        for j, item2 in enumerate(os.listdir(os.path.join(filepath,item))):
            if (item2.endswith('Completeness.csv') == True or item2.endswith('Completenes.csv') == True or item2.endswith('completeness.csv') == True) and (item2.endswith('.xlsx') == False):
                df = pd.read_csv(os.path.join(filepath,item,item2))
                nm_df = df[df['MedicareId'].isin(nm_medicare_ids)]
                print(nm_df.head())
                if nm_df.shape[0] > 0:
                    result_frame.append(nm_df)
```

```python
pd.options.display.max_colwidth = 300
pd.concat(result_frame).drop_duplicates()
```

```python
result_frame = []
for i, item in enumerate(os.listdir(filepath)):
    
    if (item.endswith('.zip')  == False) and (item.endswith('xlsx') == False) and (item != 'archive'):
        for j, item2 in enumerate(os.listdir(os.path.join(filepath,item))):
            
            if (item2.endswith('PHI.csv') == True):
                print(item2)
                df = pd.read_csv(os.path.join(filepath,item,item2))
                df['measure'] = item2
                result_frame.append(df[['medicareid','dischargemonth','measure']].drop_duplicates().sort_values(['medicareid','dischargemonth']))
                
                #df = df[df['dischargemonth'].isin(keep_months)]
                #df = df[['medicareid','recordid','dischargemonth']]
                #df2 = df.groupby(['medicareid','dischargemonth']).nunique('recordid').reset_index()
                #print(df2)
                #df3 = df2.groupby(['medicareid']).mean('recordid').reset_index()
                #print(df3[df3['medicareid'].isin(['149916','140116'])][['medicareid','recordid']])
```
